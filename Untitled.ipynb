```python
import pickle
```

```python
accidents = pickle.load( open( "filtered_accidents_df", "rb" ) )
accidents.to_csv('accidents.csv', sep=',')
```

```python
accidents
```

```python
accidents['is_1'] =[1 if len(x) == 1 else 0 for x in accidents['coordinates']]
```

```python
total_accidents = []
for index,row in accidents.iterrows():
    for cor in row['coordinates']:
        total_accidents.append((row['date'],row['date_time'],cor[0],cor[1]))
```

```python
import pandas as pd
df = pd.DataFrame(total_accidents, columns =['date', 'date_time', 'lat','lon']) 
```

```python
len(df)
```

```python
df.to_csv('accidents.csv', sep=',')
```

```python
events = pickle.load( open( "events_tweet_df", "rb" ) )
events.to_csv('events.csv', sep=',')
```
